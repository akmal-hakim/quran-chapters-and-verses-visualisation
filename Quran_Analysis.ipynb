## Quran Chapter and Verses Visualisation: Exploring the graph of Quran and more!
***
The dataset is from [The Quran Dataset](https://www.kaggle.com/datasets/imrankhan197/the-quran-dataset/).

The idea of this visualisation came when I watched a video of someone plotting a graph between Chapters and Number of Verses in the Quran. I am interested to see whether it is true or not the graphs will create "Allah" names in it. 

- Q1: Which countries that organise the most Olympic in history?


```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('The Quran Dataset.csv')
```

```python
df.head()
```

```python
#Number of Surah & ayah
print(f"Number of Surah: {len(df['surah_name_en'].unique())}")
print("Number of Ayah: " + str(len(df)))
```

***
The number of Surahs and Ayah in this dataset is accurate. We can now move further for our analysis.

***

```python
# Extract surah numbers and the total number of ayahs (verses) in each Surah
surah_numbers = df['surah_no'].unique()  # Unique surah numbers
verses_per_surah = df.groupby('surah_no')['total_ayah_surah'].first().values
```

```python
# Create a figure
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the scatter for all points to show the context
ax.scatter(surah_numbers, verses_per_surah, color='gray', s=20, label='All Verses', alpha=0.5)

# Set labels and titles
ax.set_xlabel('Surah Number', fontsize=14, color='green')
ax.set_ylabel('Ayah Number', fontsize=14, color='green')

# Set the limits of the axes
ax.set_xlim(120, 0)  # Flip the x-axis
ax.set_ylim(0, 300)

# Add a legend
ax.legend(loc='upper left')

# Add gridlines
ax.grid(True)

#Marking for Graph 1 
# Select Specific surah and filter the DataFrame, ensuring no duplicates
specific_surah_numbers_1 = [1, 2, 7, 6, 1]
specific_verses_per_surah_1 = [df[df['surah_no'] == surah]['total_ayah_surah'].values[0] for surah in specific_surah_numbers_1]

ax.plot(specific_surah_numbers_1, specific_verses_per_surah_1, color='blue', linewidth=2, label="First Specific Verses", marker='o')
ax.fill_between(specific_surah_numbers_1, specific_verses_per_surah_1, color='blue', alpha=0.2, label="Shaded Area 1")

#Marking for Graph 2 
# Select Specific surah and filter the DataFrame, ensuring no duplicates
specific_surah_numbers_2 = [5, 26, 20, 23, 27, 28, 33, 37, 43, 55, 56, 74, 77, 79, 80, 83, 89, 96, 100, 101, 102, 104, 106, 107, 109, 111, 114, 113, 112, 110, 108, 106, 105, 103, 97, 95, 94, 73, 66, 65, 63, 62, 61, 60, 49, 32, 31, 13, 8, 5]
specific_verses_per_surah_2 = [df[df['surah_no'] == surah]['total_ayah_surah'].values[0] for surah in specific_surah_numbers_2]

ax.plot(specific_surah_numbers_2, specific_verses_per_surah_2, color='green', linewidth=2, label="First Specific Verses", marker='o')
ax.fill_between(specific_surah_numbers_2, specific_verses_per_surah_2, color='green', alpha=0.2, label="Shaded Area 1")

# marker='o' will color the dots, while s='100' stands for size = 100. 

# Show the plot
plt.show()
```

***
The graph depicts the number of ayahs in each Surah, arranged by Surah number. Among these, specific Surahs are selected and plotted to illustrate a pattern. Remarkably, the shape formed by these selected Surahs visually resembles the Arabic calligraphy of the word "Allah." This unique observation suggests that by carefully choosing Surah numbers and their corresponding ayah counts, one can create meaningful shapes or patterns, in this case symbolizing a powerful religious representation.

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool

# Ensure plots show up in JupyterLab
output_notebook()

# Extract surah numbers and the total number of ayahs (verses) in each Surah
surah_numbers = df['surah_no'].unique()  # Unique surah numbers
verses_per_surah = df.groupby('surah_no')['total_ayah_surah'].first().values

# First specific line: Surah numbers [1, 2, 7, 6, 1]
specific_surah_numbers_1 = [1, 2, 7, 6, 1]
# Extract the corresponding verses for these Surah numbers
specific_verses_per_surah_1 = [df[df['surah_no'] == surah]['total_ayah_surah'].values[0] for surah in specific_surah_numbers_1]

# Second specific line: Surah numbers [5, 26, 20, 23, 27, 28, 33, 37, 43, 55, 56, 74, 77, 79, 80, 83, 89, 96, 104, 107, 109, 114, 110, 108, 103, 97, 95, 94, 73, 66, 65, 63, 62, 60, 49, 32, 31, 13, 8, 5]
#specific_surah_numbers_2 = [5, 26, 20, 23, 27, 28, 33, 37, 43, 55, 56, 74, 77, 79, 80, 83, 89, 96, 104, 107, 109, 114, 110, 108, 103, 97, 95, 94, 73, 66, 65, 63, 62, 60, 49, 32, 31, 13, 8, 5]
specific_surah_numbers_2 = [5, 26, 20, 23, 27, 28, 33, 37, 43, 55, 56, 74, 77, 79, 80, 83, 89, 96, 100, 101, 102, 104, 106, 107, 109, 111, 114, 113, 112, 110, 108, 106, 105, 103, 97, 95, 94, 73, 66, 65, 63, 62, 61, 60, 49, 32, 31, 13, 8, 5]

# Extract the corresponding verses for these Surah numbers
specific_verses_per_surah_2 = [df[df['surah_no'] == surah]['total_ayah_surah'].values[0] for surah in specific_surah_numbers_2]

# Create a Bokeh figure with customized size
p = figure(title='Qur\'an Surah Verses', x_axis_label='Surah Number', y_axis_label='Ayah Number', x_range=(114, 1), y_range=(0, 300),
           width=1000, height=600)
p.scatter(surah_numbers, verses_per_surah, size=8, color="gray", legend_label="All Verses", fill_alpha=0.5)

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Surah Number", "@x"), ("Ayah Number", "@y")]
p.add_tools(hover)

# Shade the area inside specific_surah_numbers_1
p.patch(specific_surah_numbers_1, specific_verses_per_surah_1, color="blue", fill_alpha=0.2, legend_label="Shaded Area 1")

# Shade the area inside specific_surah_numbers_2
p.patch(specific_surah_numbers_2, specific_verses_per_surah_2, color="green", fill_alpha=0.2, legend_label="Shaded Area 2")

# Add the first specific line plot
p.line(specific_surah_numbers_1, specific_verses_per_surah_1, line_width=2, color="blue", legend_label="First Specific Verses", line_dash="solid")
p.circle(specific_surah_numbers_1, specific_verses_per_surah_1, size=10, color="blue", legend_label="First Specific Verses")

# Add the second specific line plot for the new Surah numbers
p.line(specific_surah_numbers_2, specific_verses_per_surah_2, line_width=2, color="green", legend_label="New Specific Verses", line_dash="solid")
p.circle(specific_surah_numbers_2, specific_verses_per_surah_2, size=10, color="green", legend_label="New Specific Verses")

# Show the plot
show(p)
```
