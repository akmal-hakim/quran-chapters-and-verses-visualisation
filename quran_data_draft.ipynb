```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool

# Ensure plots show up in JupyterLab
output_notebook()

# Load the dataset
df = pd.read_csv('The Quran Dataset.csv')

# Extract surah numbers and the total number of ayahs (verses) in each Surah
surah_numbers = df['surah_no'].unique()  # Unique surah numbers
verses_per_surah = df.groupby('surah_no')['total_ayah_surah'].first().values

# First specific line: [1, 2, 7, 6, 1]
specific_values_1 = [1, 2, 7, 6, 1]

# Safely find the indices of the specific values
indices_1 = []
for v in specific_values_1:
    idx = np.where(verses_per_surah == v)
    if len(idx[0]) > 0:  # Check if a match was found
        indices_1.append(idx[0][0])

# Only proceed if valid indices are found
if len(indices_1) > 0:
    specific_surah_numbers_1 = surah_numbers[indices_1]
    specific_verses_per_surah_1 = verses_per_surah[indices_1]
else:
    specific_surah_numbers_1 = []
    specific_verses_per_surah_1 = []

# Second specific line: Surah numbers [4, 7, 26, 37, 4]
specific_surah_numbers_2 = [5, 26, 20, 23, 27, 28, 33, 37, 43, 55, 56, 74, 77, 79, 80, 83, 89, 96, 104, 107, 109, 114, 110, 108, 103, 97, 95, 94, 73, 66, 65, 63, 62, 60, 49, 32, 31, 13, 8, 5]

# Group and append the verse count for Surah 4
specific_verses_per_surah_2 = df[df['surah_no'].isin(specific_surah_numbers_2)].groupby('surah_no')['total_ayah_surah'].first().values
specific_verses_per_surah_2 = np.append(specific_verses_per_surah_2, specific_verses_per_surah_2[0])  # Append the verse count for Surah 4

# Create a Bokeh figure with customized size
p = figure(title='Qur\'an Surah Verses', x_axis_label='Surah Number', y_axis_label='Ayah Number', x_range=(114, 1), y_range=(0, 300),
           width=1200, height=600)
p.scatter(surah_numbers, verses_per_surah, size=8, color="gray", legend_label="All Verses", fill_alpha=0.5)

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Surah Number", "@x"), ("Ayah Number", "@y")]
p.add_tools(hover)

# Add the first specific line plot (only if valid data exists)
if len(specific_surah_numbers_1) > 0:
    p.line(specific_surah_numbers_1, specific_verses_per_surah_1, line_width=2, color="blue", legend_label="First Specific Verses", line_dash="solid")
    p.circle(specific_surah_numbers_1, specific_verses_per_surah_1, size=10, color="blue", legend_label="First Specific Verses")

# Add the second specific line plot
p.line(specific_surah_numbers_2, specific_verses_per_surah_2, line_width=2, color="green", legend_label="Second Specific Verses", line_dash="solid")
p.circle(specific_surah_numbers_2, specific_verses_per_surah_2, size=10, color="green", legend_label="Second Specific Verses")

# Show the plot
show(p)
```

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import HoverTool

# Ensure plots show up in JupyterLab
output_notebook()

# Load the dataset
df = pd.read_csv('The Quran Dataset.csv')

# Extract surah numbers and the total number of ayahs (verses) in each Surah
surah_numbers = df['surah_no'].unique()  # Unique surah numbers
verses_per_surah = df.groupby('surah_no')['total_ayah_surah'].first().values

# First specific line: Surah numbers [1, 2, 7, 6, 1]
specific_surah_numbers_1 = [1, 2, 7, 6, 1]
# Extract the corresponding verses for these Surah numbers
specific_verses_per_surah_1 = [df[df['surah_no'] == surah]['total_ayah_surah'].values[0] for surah in specific_surah_numbers_1]

# Second specific line: Surah numbers [5, 26, 20, 23, 27, 28, 33, 37, 43, 55, 56, 74, 77, 79, 80, 83, 89, 96, 104, 107, 109, 114, 110, 108, 103, 97, 95, 94, 73, 66, 65, 63, 62, 60, 49, 32, 31, 13, 8, 5]
specific_surah_numbers_2 = [5, 26, 20, 23, 27, 28, 33, 37, 43, 55, 56, 74, 77, 79, 80, 83, 89, 96, 104, 107, 109, 114, 110, 108, 103, 97, 95, 94, 73, 66, 65, 63, 62, 60, 49, 32, 31, 13, 8, 5]
# Extract the corresponding verses for these Surah numbers
specific_verses_per_surah_2 = [df[df['surah_no'] == surah]['total_ayah_surah'].values[0] for surah in specific_surah_numbers_2]

# Create a Bokeh figure with customized size
p = figure(title='Qur\'an Surah Verses', x_axis_label='Surah Number', y_axis_label='Ayah Number', x_range=(114, 1), y_range=(0, 300),
           width=1200, height=600)
p.scatter(surah_numbers, verses_per_surah, size=8, color="gray", legend_label="All Verses", fill_alpha=0.5)

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Surah Number", "@x"), ("Ayah Number", "@y")]
p.add_tools(hover)

# Shade the area inside specific_surah_numbers_1
p.patch(specific_surah_numbers_1, specific_verses_per_surah_1, color="blue", fill_alpha=0.2, legend_label="Shaded Area 1")

# Shade the area inside specific_surah_numbers_2
p.patch(specific_surah_numbers_2, specific_verses_per_surah_2, color="green", fill_alpha=0.2, legend_label="Shaded Area 2")

# Add the first specific line plot
p.line(specific_surah_numbers_1, specific_verses_per_surah_1, line_width=2, color="blue", legend_label="First Specific Verses", line_dash="solid")
p.circle(specific_surah_numbers_1, specific_verses_per_surah_1, size=10, color="blue", legend_label="First Specific Verses")

# Add the second specific line plot for the new Surah numbers
p.line(specific_surah_numbers_2, specific_verses_per_surah_2, line_width=2, color="green", legend_label="New Specific Verses", line_dash="solid")
p.circle(specific_surah_numbers_2, specific_verses_per_surah_2, size=10, color="green", legend_label="New Specific Verses")

# Show the plot
show(p)
```
